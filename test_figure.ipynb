```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('classic')
%matplotlib inline
from matplotlib import cycler
colors = cycler('color',
                ['#EE6666', '#3388BB', '#9988DD',
                 '#EECC55', '#88BB44', '#FFBBBB'])
# define overall axes
plt.rc('axes', facecolor='#E6E6E6', edgecolor='#3388BB',
       axisbelow = True, grid = True, prop_cycle = colors)
plt.rc('grid', color = 'w', linestyle = 'solid')
# design x-axis and y-axis
plt.rc('xtick', direction = 'out', color = 'gray')
plt.rc('ytick', direction = 'out', color = 'gray')
plt.rc('patch', edgecolor = '#E6E6E6') # how edges looks like
```

```python
x = np.random.randn(1000)
plt.hist(x)
```

```python
for i in range(4):
  plt.plot(np.random.rand(10))
```

```python
y = np.linspace(0,10,100)

plt.plot(y, np.sin(y))
plt.plot(y, np.cos(y))

plt.show()
```

```python
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(bars))

plt.bar(y_pos, height) # create bar
plt.xticks(y_pos, bars) # create name
plt.show()
```

```python
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')

fig1, ax1 = plt.subplots() # allow plot pie
ax1.pie(height,
        labels=bars,
        autopct = '%1.1f%%',
        startangle = 90)
ax1.axis('equal') # equal aspect ratio
plt.show()
```
